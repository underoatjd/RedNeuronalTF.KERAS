```python
import numpy as np
import pandas as pd

datos = pd.read_csv('celcius.csv')
datos
```

```python
x = datos["celcius"].values
y = datos["fahrenheit"].values
```

```python
import tensorflow as tf
modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)), # La entrada, los celcius.
    tf.keras.layers.Dense(units=1) # Salida de datos, o sea los Fahrenheit.
])

modelo.compile(
    optimizer=tf.keras.optimizers.Adam(.1),
    loss="mean_squared_error"
)
```

```python
entrenamiento = modelo.fit(x, y, epochs=1000)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epocas")
plt.ylabel("Magnitud Perdida")
plt.plot(entrenamiento.history['loss'])
```

```python
celcius = 30

predicion = modelo.predict([celcius])
print(f"La predicción de {celcius} grados Celsius es igual a {predicion[0]} grados en Fahrenheit")
```
